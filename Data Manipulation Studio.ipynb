## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

balt_df = pd.read_csv('baltimore_9-24-2016_9-30-2017.csv')
bost_df = pd.read_csv('boston_9-24-2016_9-30-2017.csv')
york_df = pd.read_csv('new-york_9-24-2016_9-30-2017.csv')
phil_df = pd.read_csv('philadelphia_9-24-2016_9-30-2017.csv')
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
for col in balt_df.columns:
    pct_missing = np.mean(balt_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
print('')
    
for col in bost_df.columns:
    pct_missing = np.mean(bost_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
print('')
    
for col in york_df.columns:
    pct_missing = np.mean(york_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
print('')
    
for col in phil_df.columns:
    pct_missing = np.mean(phil_df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
balt_df = balt_df.drop(["Commodity Name", "Type", "Grade","Origin District", "Environment", "Quality", "Condition", "Appearance", "Storage","Crop", "Trans Mode", "Color", "Sub Variety"], axis=1) 
bost_df = bost_df.drop(["Commodity Name", "Type", "Grade","Origin District", "Environment", "Quality", "Condition", "Appearance", "Storage","Crop", "Trans Mode", "Color", "Sub Variety"], axis=1)
york_df = york_df.drop(["Commodity Name", "Type", "Grade","Origin District", "Environment", "Quality", "Condition", "Appearance", "Storage","Crop", "Trans Mode", "Color", "Sub Variety"], axis=1)
phil_df = phil_df.drop(["Commodity Name", "Type", "Grade","Origin District", "Environment", "Quality", "Condition", "Appearance", "Storage","Crop", "Trans Mode", "Color", "Sub Variety"], axis=1)
```

```python
balt_df.head()
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!

northeast_df = pd.concat([balt_df, york_df, bost_df, phil_df],axis=0)
northeast_df.head()
```
